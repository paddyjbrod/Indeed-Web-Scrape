```python
## functions for scraping different elements
def extract_location_from_resultRow(result):
    try:
        location= (result.find('span', {'class': 'location'})).text.strip()
    except:
        location = ''
    return location

def extract_company_from_resultRow(result):
    try:
        company = (result.find('span', {'class':'company'})).text.strip()
    except:
        company = ''
    return company

def extract_rating_from_resultRow(result):
    try:
        rating= (result.find("div", {'class':"sjcl"})).text.strip()
    except:
        rating= ''
    return rating

def extract_title_from_resultRow(result):
    try:
        title = (result.find('h2', {'class':'title'})).text.strip()
        title_text = title
    except: 
        title_text = ''
    return title_text

def extract_salary_from_resultRow(result):
    try:
        salary = (result.find('span', {'class' : 'salary no-wrap'})).text.strip()
    except:
        salary = ''
        no_salary.append(salary)
    salary_text = salary
    return salary_text

def extract_summary_from_resultRow(result):
    try: 
        summary = (result.find('div', {'class':'summary'})).text.strip()
    except: 
        summary = ''
    return summary

def extract_date_from_resultRow(result):
    try: 
        date = (result.find(class_='date').text.strip(' ago').strip())
    except: 
        date = ''
    return date
```

## SYDNEY DATA ANALYST SCRAPE

```python
CITY=['sydney']
URL_SEARCH_TERM = 'data analyst'
```

```python
# scraping sydney data analyst
    
for city in CITY:
    job_dict = []
    now = dt.datetime.now()
    for start in range(10, 1000,10):
        
#         URL = "https://au.indeed.com/jobs?q=data&l="+urllib.quote(city)"&start="+(start)
        URL = "https://au.indeed.com/jobs?q="+urllib.parse.quote(URL_SEARCH_TERM)+"&l="+urllib.parse.quote(city)+"&start="+str(start)
        
        driver.get(URL)
        soup = BeautifulSoup(driver.page_source, "lxml")

        for i in soup.findAll("div", {"class" : "jobsearch-SerpJobCard unifiedRow row result clickcard"}):
            
            location = extract_location_from_resultRow(i)
            company = extract_company_from_resultRow(i)
            summary = extract_summary_from_resultRow(i)
            rating=extract_rating_from_resultRow(i)
            title = extract_title_from_resultRow(i)
            salary = extract_salary_from_resultRow(i)
            post_date = extract_date_from_resultRow(i)

            job_dict.append([location, company, summary, title, salary, post_date,now, rating])
    job_df = pd.DataFrame(job_dict, columns=['location', 'company', 'summary',  'title', 'salary', 'post_date','pull_date','rating']) 
```
